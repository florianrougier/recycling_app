# Checking out the data/images


First we have our dependencies

```
import os
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

Now this is where we setup the directories

```
base_dir = './dataset-resized'
cardboard_dir = base_dir + '/cardboard'
glass_dir = base_dir + '/glass'
metal_dir = base_dir + '/metal'
paper_dir = base_dir + '/paper'
plastic_dir = base_dir + '/plastic'
```

We could spend time showing the images 1 by 1 like this, but that would take a long time.

```
img = mpimg.imread(glass_dir + '/glass17.jpg')
imgplot = plt.imshow(img)
plt.show()
```

So instead, we make this function, that takes 6 random images from the supplied directory, and plots them. (You can skip the whole labeled=False thing for now)

```
def plot_images(path, labeled=False, max_images=6):
  files = os.listdir(path)
  random.shuffle(files)
  amount = 0
  fig = plt.figure(figsize=(12, 6))

  for file in files:
    if file.endswith('.jpg'):
      if amount == max_images:
        break

      img = mpimg.imread(os.path.join(path, file))
      plt.subplot(231+amount)
      if labeled:
        plt.title(file.split('_')[1])
      imgplot = plt.imshow(img)

      amount += 1
```
